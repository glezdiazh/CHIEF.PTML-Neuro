```python
import itertools

# -----------------------------
# COLUMNA BASE
# -----------------------------
base_col = 'fchiral'

# -----------------------------
# COLUMNAS PARA COMBINAR
# -----------------------------
interaction_cols = [
    'V2=ALOGP',
    'V3=PSA',
    'V4=NVLR'
]

# -----------------------------
# COMPROBACIÓN DE SEGURIDAD
# -----------------------------
missing = [c for c in [base_col] + interaction_cols if c not in df.columns]
if missing:
    raise ValueError(f"Columnas no encontradas en df: {missing}")

# -----------------------------
# GENERAR TODAS LAS COMBINACIONES
# -----------------------------
new_features = []

for r in range(1, len(interaction_cols) + 1):
    for combo in itertools.combinations(interaction_cols, r):

        # limpiar nombres para evitar problemas
        clean_combo = [
            c.replace('(', '')
             .replace(')', '')
             .replace(',', '')
             .replace(' ', '')
            for c in combo
        ]

        base_clean = (
            base_col.replace('(', '')
                    .replace(')', '')
                    .replace(',', '')
                    .replace(' ', '')
        )

        new_col_name = base_clean + '_x_' + '_x_'.join(clean_combo)

        # cálculo de la nueva feature
        df[new_col_name] = df[base_col]
        for c in combo:
            df[new_col_name] *= df[c]

        new_features.append(new_col_name)

# -----------------------------
# RESULTADO
# -----------------------------
print(f"✔ Se crearon {len(new_features)} nuevas columnas:")
for col in new_features:
    print(col)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, roc_auc_score, roc_curve, 
                             confusion_matrix, ConfusionMatrixDisplay)

print("Librerías importadas ✅")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

print("Modelos listos ✅")
```

```python
models = {
    "LDA": LinearDiscriminantAnalysis(),
}

param_grids = {
    "LDA": {"solver": ["svd", "lsqr"]},
}

print("Modelos y grids preparados ✅")
```

```python
models = {
    "RandomForest": RandomForestClassifier(random_state=42),
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=42),
    "LinearSVM": LinearSVC(random_state=42),
    "KNN": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "SGDClassifier": SGDClassifier(random_state=42)
}

param_grids = {
    "RandomForest": {"n_estimators": [50, 100], "max_depth": [None, 5], "min_samples_split": [2]},
    "LogisticRegression": {"C": [0.1, 1], "penalty": ["l2"]},
    "LinearSVM": {"C": [0.1, 1]},
    "KNN": {"n_neighbors": [3, 5], "weights": ["uniform"]},
    "GaussianNB": {},
    "DecisionTree": {"max_depth": [None, 5], "min_samples_split": [2, 5]},
    "SGDClassifier": {"loss": ["hinge", "log_loss"], "penalty": ["l2", "l1"], "alpha": [0.0001, 0.001]}
}

print("Modelos y grids preparados ✅")
```

```python
# Validación cruzada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Listas y diccionarios para resultados y gráficos
results = []
train_probas = {}
test_probas = {}
best_estimators = {}

# Carpeta para guardar todo (tablas + gráficos)
output_dir = "resultados_modelos"
os.makedirs(output_dir, exist_ok=True)
print(f"Carpeta creada: {output_dir} 📁")
```

```python
# Asegúrate de tener este import al inicio del notebook (¡IMPORTANTE!)
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Tu bucle for MODIFICADO (copia y pega esto completo)
for model_name in models.keys():
    print(f"\nEntrenando {model_name}... ⏳")
   
    model = models[model_name]
    grid = param_grids[model_name]
   
    gs = GridSearchCV(
        estimator=model,
        param_grid=grid,
        scoring='f1',
        cv=cv,
        n_jobs=-1,
        verbose=1
    )
   
    gs.fit(X_train, y_train)
    best_model = gs.best_estimator_
    best_estimators[model_name] = best_model # Guardamos para gráficos posteriores
   
    # Predicciones
    y_train_pred = best_model.predict(X_train)
    y_test_pred = best_model.predict(X_test)
   
    # Probabilidades / scores para ROC
    if hasattr(best_model, "predict_proba"):
        y_train_proba = best_model.predict_proba(X_train)[:, 1]
        y_test_proba = best_model.predict_proba(X_test)[:, 1]
    elif hasattr(best_model, "decision_function"):
        y_train_proba = best_model.decision_function(X_train)
        y_test_proba = best_model.decision_function(X_test)
    else:
        y_train_proba = None
        y_test_proba = None
   
    train_probas[model_name] = y_train_proba
    test_probas[model_name] = y_test_proba
    
    # 🆕 NUEVO: Calcula Specificity (Sp) usando matriz de confusión
    cm_train = confusion_matrix(y_train, y_train_pred)
    cm_test = confusion_matrix(y_test, y_test_pred)
    
    # Specificity: TN / (TN + FP) para clasificación binaria (0=negativo, 1=positivo)
    if len(cm_train) == 2:  # Binario
        sp_train = cm_train[0, 0] / (cm_train[0, 0] + cm_train[0, 1]) if (cm_train[0, 0] + cm_train[0, 1]) > 0 else 0
        sp_test = cm_test[0, 0] / (cm_test[0, 0] + cm_test[0, 1]) if (cm_test[0, 0] + cm_test[0, 1]) > 0 else 0
    else:
        sp_train = np.nan  # Multiclase: usa average='macro' si necesitas
        sp_test = np.nan
    
    # Métricas (AHORA INCLUYE Specificity)
    metrics = {
        "Model": model_name,
        "Best_Params": gs.best_params_,
        "Train_Accuracy": accuracy_score(y_train, y_train_pred),
        "Test_Accuracy": accuracy_score(y_test, y_test_pred),
        "Train_Precision": precision_score(y_train, y_train_pred),
        "Test_Precision": precision_score(y_test, y_test_pred),
        "Train_Recall": recall_score(y_train, y_train_pred),      # ← Sensitivity (Sn)
        "Test_Recall": recall_score(y_test, y_test_pred),         # ← Sensitivity (Sn)
        "Train_Specificity": sp_train,                            # 🆕 Sp TRAIN
        "Test_Specificity": sp_test,                              # 🆕 Sp TEST
        "Train_F1": f1_score(y_train, y_train_pred),
        "Test_F1": f1_score(y_test, y_test_pred),
        "Train_ROC_AUC": roc_auc_score(y_train, y_train_proba) if y_train_proba is not None else np.nan,
        "Test_ROC_AUC": roc_auc_score(y_test, y_test_proba) if y_test_proba is not None else np.nan
    }
   
    results.append(metrics)
    print(f"{model_name} terminado ✅ (Test F1: {metrics['Test_F1']:.4f}, Test Sp: {metrics['Test_Specificity']:.4f})")
```

```python
# ────────────────────────────────────────────────
# Generar predicciones y unirlas al Excel grande usando "Fila ID"
# ────────────────────────────────────────────────

import pandas as pd
import numpy as np

# 1. Carga el Excel GRANDE (el que tiene todas las columnas originales + "Fila ID")
df_grande = pd.read_excel('/kaggle/input/chiefptml-leish-ok/SI01-PTML_Leishmania_con_fchiral.xlsx')          # ← CAMBIA este nombre por el real

# Verificación básica
if 'Fila ID' not in df_grande.columns:
    print("¡ERROR! No se encuentra la columna 'Fila ID' en el Excel grande")
    # Puedes parar aquí si quieres: raise ValueError("Falta 'Fila ID'")
else:
    print(f"Excel grande cargado. Filas: {len(df_grande)}, 'Fila ID' encontrado ✓")

# 2. Creamos un DataFrame solo con las predicciones + "Fila ID"
predicciones_df = pd.DataFrame()
predicciones_df['Fila ID'] = list(id_train) + list(id_test)

# Recorremos cada modelo y añadimos sus predicciones y probabilidades
for model_name in best_estimators.keys():
    best_model = best_estimators[model_name]
    
    # Predicciones
    preds_train = best_model.predict(X_train)
    preds_test  = best_model.predict(X_test)
    predicciones_df[f'pred_{model_name}'] = np.concatenate([preds_train, preds_test])
    
    # Probabilidades (solo si existen)
    proba_train = train_probas.get(model_name)
    proba_test  = test_probas.get(model_name)
    
    if proba_train is not None and proba_test is not None:
        predicciones_df[f'proba_{model_name}'] = np.concatenate([proba_train, proba_test])
    else:
        print(f"Nota: {model_name} no tiene predict_proba ni decision_function → sin columna proba")

# 3. Unimos (merge) por "Fila ID" al Excel grande
df_final = df_grande.merge(
    predicciones_df,
    on='Fila ID',
    how='left'          # conserva todas las filas del Excel grande
)

# 4. Verificaciones útiles
print(f"\nFilas en Excel grande original: {len(df_grande)}")
print(f"Filas con predicciones generadas: {len(predicciones_df)}")
print(f"Filas en el resultado final: {len(df_final)}")
print(f"Predicciones añadidas a {predicciones_df['Fila ID'].isin(df_grande['Fila ID']).sum()} filas")

# Muestra algunas filas para chequear
print("\nPrimeras 5 filas del resultado (con predicciones):")
display(df_final.head())

print("\nÚltimas 5 filas:")
display(df_final.tail())

# 5. Guardamos el Excel ampliado
nombre_salida = 'excel_grande_con_predicciones.xlsx'
df_final.to_excel(nombre_salida, index=False)
print(f"\n¡Guardado exitoso! Archivo creado: {nombre_salida}")
```

```python
def plot_roc_multi(y_true, probas_dict, title, filename):
    plt.figure(figsize=(10, 8))
    for model_name, y_proba in probas_dict.items():
        if y_proba is not None:
            fpr, tpr, _ = roc_curve(y_true, y_proba)
            auc = roc_auc_score(y_true, y_proba)
            plt.plot(fpr, tpr, label=f"{model_name} (AUC: {auc:.4f})")
    
    plt.plot([0, 1], [0, 1], 'k--', label='Aleatorio')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.grid(True)
    
    save_path = f"{output_dir}/{filename}"
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Guardado: {save_path}")

plot_roc_multi(y_train, train_probas, "Curvas ROC - Train", "roc_train.png")
plot_roc_multi(y_test, test_probas, "Curvas ROC - Test", "roc_test.png")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Nombres de features (esto ya lo tienes bien)
feature_names = (X_train.columns
                 if hasattr(X_train, "columns")
                 else [f"Feature_{i}" for i in range(X_train.shape[1])])

# ────────────────────────────────────────────────
# Solo un modelo: LDA
# ────────────────────────────────────────────────

model_name = "LDA"
model = best_estimators[model_name]   # tu mejor modelo LDA después de GridSearch

# Verificamos si tiene coef_ (LDA sí lo tiene)
if not hasattr(model, "coef_"):
    print(f"{model_name} no tiene coef_ → No se puede calcular importancia")
else:
    # Para LDA: tomamos valor absoluto de los coeficientes
    # Si es binario → coef_.shape = (1, n_features)
    # Si es multiclase → coef_.shape = (n_clases, n_features) → promediamos o tomamos abs y max por feature
    coef = model.coef_
    
    # Caso común: binario → coef_ es (1, n_features)
    if coef.ndim == 2 and coef.shape[0] == 1:
        importances = np.abs(coef[0])           # valor absoluto
    else:
        # Multiclase: tomamos el valor absoluto promedio por feature
        importances = np.abs(coef).mean(axis=0)   # o .max(axis=0) si prefieres el más discriminante
    
    # Top 15 features
    indices = np.argsort(importances)[::-1][:15]
    top_feat = np.array(feature_names)[indices]
    top_imp = importances[indices]
    
    # Gráfico
    plt.figure(figsize=(10, 8))
    sns.barplot(x=top_imp, y=top_feat, palette="viridis")
    plt.title(f"Top 15 Feature Importance - {model_name} (basado en |coef_|)")
    plt.xlabel("Importancia (valor absoluto de coeficientes)")
    plt.tight_layout()
    
    # Guardar
    path = f"{output_dir}/feature_importance_{model_name}.png"
    plt.savefig(path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Guardado: {path}")
```
